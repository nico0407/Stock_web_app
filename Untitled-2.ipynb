```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, kpss, acf, pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import het_arch
from arch import arch_model
import seaborn as sns


# Generate 10 sequential dates starting from today
dates = [datetime.today() - timedelta(days=i) for i in range(365)]

# Generate random volume values
volumes = np.random.randint(100, 1000, size=365)

# Create the DataFrame
FX_Data = pd.DataFrame({
    'date': dates,
    'volume': volumes
})

FX_Data
```

# GPT

```python
# Ensure 'date' is datetime and set as index
FX_Data['date'] = pd.to_datetime(FX_Data['date'])
FX_Data.set_index('date', inplace=True)
FX_Data.sort_index(inplace=True)

# Plot raw data
FX_Data['volume'].plot(title='FX Swap Contract Volume')
plt.show()

# 1. Stationarity Test (ADF and KPSS)
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print("ADF Test Statistic:", result[0])
    print("p-value:", result[1])
    return result[1]

def kpss_test(series):
    result = kpss(series, regression='c')
    print("KPSS Test Statistic:", result[0])
    print("p-value:", result[1])
    return result[1]

print("ADF Test:")
adf_test(FX_Data['volume'])
print("KPSS Test:")
kpss_test(FX_Data['volume'])

# 2. De-trending and Seasonality Extraction
decomp = seasonal_decompose(FX_Data['volume'], model='additive', period=30)  # or 365 for yearly seasonality
decomp.plot()
plt.show()

# De-trended = volume - trend
detrended = FX_Data['volume'] - decomp.trend
deseasonalized = detrended.dropna()

# 3. Differencing to check integration (if necessary)
diff_series = FX_Data['volume'].diff().dropna()

print("ADF on first difference:")
adf_test(diff_series)
print("KPSS on first difference:")
kpss_test(diff_series)

# 4. Autocorrelation and Partial Autocorrelation
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(diff_series, lags=30)
plt.title('ACF of Differenced Series')
plt.show()

plot_pacf(diff_series, lags=30)
plt.title('PACF of Differenced Series')
plt.show()

# 5. ARIMA model (optional fitting)
model = ARIMA(FX_Data['volume'], order=(1,1,1))  # Adjust p,d,q after ACF/PACF
fit_model = model.fit()
print(fit_model.summary())

# 6. Heteroskedasticity (ARCH test)
arch_test = het_arch(fit_model.resid.dropna())
print("ARCH Test for Heteroskedasticity:")
print(f"F-statistic: {arch_test[0]}, p-value: {arch_test[1]}")

# 7. Fit GARCH model if heteroskedasticity is present
if arch_test[1] < 0.05:
    garch_model = arch_model(fit_model.resid.dropna(), vol='Garch', p=1, q=1)
    garch_fit = garch_model.fit(disp="off")
    print(garch_fit.summary())
    garch_fit.plot()
    plt.show()
```

```python
# Set style for plots
#plt.style.use('seaborn')
sns.set_palette("husl")

# Assuming FX_Data is already loaded as a pandas DataFrame
# If not, you would load it with:
# FX_Data = pd.read_csv('your_file.csv', parse_dates=['date'], index_col='date')

# Ensure date is the index and sorted
FX_Data = FX_Data.set_index('date').sort_index()

# Basic data inspection
print(FX_Data.head())
print("\nData info:")
print(FX_Data.info())
print("\nSummary statistics:")
print(FX_Data.describe())

# Plot the raw time series
plt.figure(figsize=(12, 6))
FX_Data['volume'].plot(title='Daily FX Swap Trading Volume')
plt.ylabel('Volume')
plt.xlabel('Date')
plt.grid(True)
plt.show()

## 1. Trend Analysis and De-trending
# Decompose the time series to observe trend and seasonality
# Using additive model first (multiplicative if variance grows with level)
decomposition = seasonal_decompose(FX_Data['volume'], model='additive', period=5)  # Adjust period as needed

# Plot decomposition
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.suptitle('Time Series Decomposition (Additive Model)')
plt.tight_layout()
plt.show()

# Detrending by differencing
FX_Data['volume_diff'] = FX_Data['volume'].diff()

# Plot detrended series
plt.figure(figsize=(12, 6))
FX_Data['volume_diff'].plot(title='First Difference of FX Swap Volume (Detrended)')
plt.ylabel('Volume Difference')
plt.grid(True)
plt.show()

## 2. Stationarity Tests
# ADF Test (null hypothesis: unit root exists - non-stationary)
adf_result = adfuller(FX_Data['volume'].dropna())
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'   {key}: {value}')

# KPSS Test (null hypothesis: stationary)
kpss_result = kpss(FX_Data['volume'].dropna())
print(f'\nKPSS Statistic: {kpss_result[0]}')
print(f'p-value: {kpss_result[1]}')
print('Critical Values:')
for key, value in kpss_result[3].items():
    print(f'   {key}: {value}')

## 3. Seasonality Analysis
# Autocorrelation and Partial Autocorrelation plots
plt.figure(figsize=(12, 6))
plot_acf(FX_Data['volume'].dropna(), lags=40, title='Autocorrelation of FX Swap Volume')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(FX_Data['volume'].dropna(), lags=40, title='Partial Autocorrelation of FX Swap Volume')
plt.show()

# Seasonal decomposition with different periods to identify cycles
for period in [5, 20, 30, 60]:  # Trying different potential seasonal periods
    try:
        decomposition = seasonal_decompose(FX_Data['volume'].dropna(), model='additive', period=period)
        plt.figure(figsize=(12, 8))
        decomposition.plot()
        plt.suptitle(f'Seasonal Decomposition with Period={period}')
        plt.tight_layout()
        plt.show()
    except:
        print(f"Could not decompose with period {period}")

## 4. ARIMA Properties
# Based on ACF/PACF, we can identify potential AR and MA terms
# Let's fit an ARIMA model (order will need to be determined)
from statsmodels.tsa.arima.model import ARIMA

# Example ARIMA(1,1,1) - you should determine proper order based on ACF/PACF and tests
model = ARIMA(FX_Data['volume'], order=(1,1,1))
results = model.fit()
print(results.summary())

# Check residuals
residuals = results.resid
plt.figure(figsize=(12, 6))
residuals.plot(title='ARIMA Model Residuals')
plt.grid(True)
plt.show()

# ACF/PACF of residuals to check if model captured all structure
plt.figure(figsize=(12, 6))
plot_acf(residuals.dropna(), lags=40, title='ACF of ARIMA Residuals')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(residuals.dropna(), lags=40, title='PACF of ARIMA Residuals')
plt.show()

## 5. Heteroskedasticity Analysis
# Plot squared residuals to visualize volatility clustering
plt.figure(figsize=(12, 6))
(residuals**2).plot(title='Squared Residuals (Visualizing Volatility Clustering)')
plt.grid(True)
plt.show()

# ARCH test for heteroskedasticity
from statsmodels.stats.diagnostic import het_arch
arch_test = het_arch(residuals.dropna())
print(f'\nARCH Test p-value: {arch_test[1]}')  # Small p-value indicates ARCH effects

# Fit GARCH model to model volatility
garch_model = arch_model(residuals.dropna(), vol='Garch', p=1, q=1)
garch_results = garch_model.fit(update_freq=5)
print(garch_results.summary())

# Plot conditional volatility
plt.figure(figsize=(12, 6))
garch_results.conditional_volatility.plot(title='Conditional Volatility from GARCH Model')
plt.grid(True)
plt.show()

## 6. Rolling Statistics to Examine Time-Varying Properties
# Rolling mean and standard deviation
rolling_window = 20  # 20 trading days ~ 1 month
FX_Data['rolling_mean'] = FX_Data['volume'].rolling(window=rolling_window).mean()
FX_Data['rolling_std'] = FX_Data['volume'].rolling(window=rolling_window).std()

plt.figure(figsize=(12, 6))
plt.plot(FX_Data['volume'], label='Original')
plt.plot(FX_Data['rolling_mean'], label=f'Rolling Mean ({rolling_window} days)')
plt.plot(FX_Data['rolling_std'], label=f'Rolling Std Dev ({rolling_window} days)')
plt.legend()
plt.title('Rolling Mean & Standard Deviation')
plt.grid(True)
plt.show()
```
