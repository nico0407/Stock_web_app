\section{Introduction}
We seek to decompose a daily time series $X_t$ (``volume'') into three parts: the trend component $T_t$, the seasonal component $S_t$, and the residual~$R_t$.  The additive model is
\begin{align}
X_t = T_t + S_t + R_t.
\end{align}

\section{Trend Estimation via Moving Average}
A simple estimate of the trend is the centered moving average (MA) of window length $m=2k+1$.  Define
\begin{align}
\widehat{T}_t = \frac{1}{m}\sum_{i=-k}^{k} X_{t+i},
\end{align}
where $m=30$ days in our example ($k=14$).  In Python:
# 30-day centered moving average
df['trend'] = df['volume'].rolling(window=30, center=True).mean()
# detrended series:
df['detrended'] = df['volume'] - df['trend']

\section{Classical Seasonal Decomposition}
Given a period $m$ (here weekly, $m=7$), classical additive decomposition estimates $T_t$ by smoothing, then the seasonal component by averaging:
\begin{align}
\widehat{T}_t &= \text{MA}(X_t),\\
\widehat{S}_j &= \frac{1}{N}\sum_{\substack{t=1\dots,N\\t\equiv j\,(\mathrm{mod}\,m)}} \bigl(X_t - \widehat{T}_t\bigr),\quad j=1,\dots,m,\\
\widehat{R}_t &= X_t - \widehat{T}_t - \widehat{S}_{t\bmod m}.
\end{align}
In Python with \texttt{statsmodels}:
from statsmodels.tsa.seasonal import seasonal_decompose
decomp = seasonal_decompose(df['volume'], model='additive', period=7)
decomp.plot()


\section{STL Decomposition}
STL (\emph{Seasonal and Trend via Loess}) fits both components via local regression.  It handles irregular seasonal patterns and outliers:
from statsmodels.tsa.seasonal import STL
stl = STL(df['volume'], period=7, robust=True)
res = stl.fit()
res.plot()

\section{Autocorrelation Analysis}
Seasonality is confirmed by spikes in the autocorrelation function (ACF) at lags that are multiples of~$m$:
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df['volume'].dropna(), lags=30)

\section{Summary of Key Formulas}
\begin{align}
X_t &= T_t + S_t + R_t,\\
\widehat{T}_t &= \frac{1}{m}\sum_{i=-k}^{k} X_{t+i},\\
\widehat{S}_j &= \frac{1}{N}\sum_{t\equiv j\,(\mathrm{mod}\,m)} \bigl(X_t - \widehat{T}_t\bigr),\\
\widehat{R}_t &= X_t - \widehat{T}_t - \widehat{S}_t.
\end{align}

\section{GARCH(1,1) Modeling of Daily Volume}

We start from the raw volume series \(X_t\) and form the log‑returns
\begin{align}
r_t = \ln(X_t) - \ln(X_{t-1}),\quad t=2,\dots,T.
\end{align}

\subsection{Stationarity Check}
Perform the Augmented Dickey–Fuller test on \(\{r_t\}\):
\begin{align}
H_0:\ \text{unit root (nonstationary)}
\quad\text{vs.}\quad
H_1:\ \text{stationary}.
\end{align}
Reject \(H_0\) if the ADF \(p\)-value \(<0.05\).

\subsection{GARCH(1,1) Specification}
Model the returns via
\begin{align}
r_t &= \mu + \varepsilon_t,\quad
\varepsilon_t = \sigma_t\,z_t,\quad z_t\sim \mathcal{N}(0,1),\\
\sigma_t^2 &= \omega + \alpha\,\varepsilon_{t-1}^2 + \beta\,\sigma_{t-1}^2,
\end{align}
with parameter constraints for covariance stationarity:
\begin{align}
\omega>0,\quad \alpha\ge0,\quad \beta\ge0,\quad \alpha + \beta < 1.
\end{align}

\subsection{Unconditional Variance}
Under stationarity,
\begin{align}
\operatorname{Var}(r_t)
= \mathbb{E}\bigl[\sigma_t^2\bigr]
= \frac{\omega}{1 - \alpha - \beta}.
\end{align}

\subsection{Diagnostics}
After fitting, extract standardized residuals
\(\displaystyle z_t = \varepsilon_t/\sigma_t\)
and check:1) ACF of \(z_t\) and \(z_t^2\) (no significant lags). 2) tem Ljung–Box tests on \(z_t\) and \(z_t^2\) (\(p\)-values \(>0.05\)).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.diagnostic import acorr_ljungbox
from arch import arch_model

# 1. Load your data (adjust path/column names as needed)
df = df.asfreq('D')
df['r'] = np.log(df['volume']).diff().dropna()

# 2. Stationarity check on returns
adf_stat, pval, *_ = adfuller(df['r'].dropna(), autolag='AIC')
print(f"ADF stat = {adf_stat:.3f}, p‐value = {pval:.3f}")
# if pval < .05, OK — otherwise consider differencing again (rare for returns)

# 3. Fit AR(0)–GARCH(1,1)
rets = df['r'].dropna()
am = arch_model(rets, mean='Constant', vol='GARCH', p=1, q=1, dist='normal')
res = am.fit(update_freq=5, disp='off')
print(res.summary())

# 4. Plot log‐returns & conditional volatility
plt.figure(figsize=(10,3))
plt.plot(rets.index, rets)
plt.title("Log‑Returns")
plt.show()

plt.figure(figsize=(10,3))
plt.plot(res.conditional_volatility.index, res.conditional_volatility)
plt.title("Estimated Conditional Volatility σ_t")
plt.show()

# 5. Diagnostics on standardized residuals
std_resid = res.std_resid

#  5a. ACF of squared std‐resids
plot_acf(std_resid**2, lags=20)
plt.title("ACF of (Std Residuals)^2")
plt.show()

#  5b. Ljung–Box
lb1 = acorr_ljungbox(std_resid, lags=[10], return_df=True)
lb2 = acorr_ljungbox(std_resid**2, lags=[10], return_df=True)
print("LB test on std_resid:\n", lb1)
print("LB test on std_resid^2:\n", lb2)

# 6. Forecasting 5‐step ahead
horizon = 5
fore = res.forecast(horizon=horizon, method='analytic')

# the arch forecast object stores variances in a (T × H) DataFrame
var_fore = fore.variance.iloc[-1]            # last row, horizons 1..5
std_fore = np.sqrt(var_fore)                 # σ_{T+1},…,σ_{T+5}

# math: for GARCH(1,1) we know
#   Var(r_{T+h}|Ω_T) 
#     = ω * (1 - φ^h)/(1 - φ)  +  φ^h * σ_T^2,
#   where φ = α + β  and  σ_T^2 is the last estimated variance.
#
# 95%‐CI for r_{T+h} is  μ  ± 1.96·σ_{T+h}.  
#  (and vol forecasts converge to the long‐run var ω/(1−φ).)

# build future index
last_date = rets.index[-1]
future_idx = pd.date_range(last_date + pd.Timedelta(days=1), periods=horizon, freq='D')

# 7. Plot volatility‐forecast
plt.figure(figsize=(10,3))
# historical
plt.plot(res.conditional_volatility.index, res.conditional_volatility, label='σ_t (est)')
# forecast
plt.plot(future_idx, std_fore.values, linestyle='--', label='σ_{T+h} forecast')
plt.fill_between(future_idx,
                 std_fore.values * 0.95,   # an example “±5%” band (you can replace with analytic CI)
                 std_fore.values * 1.05,
                 alpha=0.2)
plt.title("GARCH(1,1) Volatility Forecast (5 days out)")
plt.legend()
plt.show()

# 8. (Optionally) Plot return‐forecast CIs
mu = res.params['mu']
mean_fore = np.repeat(mu, horizon)

upper = mean_fore + 1.96 * std_fore
lower = mean_fore - 1.96 * std_fore

plt.figure(figsize=(10,3))
plt.plot(future_idx, mean_fore, label='E[r_{T+h}]')
plt.fill_between(future_idx, lower, upper, alpha=0.2)
plt.title("Return Forecast ±95% CI")
plt.show()
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from arch import arch_model
import matplotlib.pyplot as plt

# 1. Load & prepare
df = df.asfreq('D')

# 2. Inspect
print(df['volume'].describe())
print(df['volume'].head(), df['volume'].tail(), sep='\n')

# 3. Filter out non-positive volumes
df = df[df['volume'] > 0].copy()

# 4. Compute log-returns
df['r'] = np.log(df['volume']).diff()

# 5. Drop NaNs and confirm length
rets = df['r'].dropna()
print(f"Number of usable returns: {len(rets)}")

if len(rets) == 0:
    raise ValueError("No positive, non-missing volumes remain after filtering. Check your data!")

# 6. ADF test for stationarity
adf_stat, pval, _, _, crit_vals, _ = adfuller(rets, autolag='AIC')
print(f"ADF Statistic: {adf_stat:.4f}")
print(f"p-value:       {pval:.4f}")
for k, v in crit_vals.items():
    print(f"Critical Value ({k}): {v:.4f}")

# 7. Fit GARCH(1,1)
am = arch_model(rets, mean='Constant', vol='GARCH', p=1, q=1, dist='normal')
res = am.fit(update_freq=5, disp='off')
print(res.summary())

# 8. Plot returns & conditional vol
plt.figure(figsize=(10,4))
plt.plot(rets.index, rets, label='Log-Returns')
plt.title('Log-Returns of Volume')
plt.show()

plt.figure(figsize=(10,4))
plt.plot(res.conditional_volatility.index,
         res.conditional_volatility,
         label='σ_t')
plt.title('Estimated Conditional Volatility')
plt.show()

# 9. Forecast 5 days out
horizon = 5
fore = res.forecast(horizon=horizon, method='analytic')
var_fore = fore.variance.iloc[-1]
std_fore = np.sqrt(var_fore)

future_idx = pd.date_range(rets.index[-1] + pd.Timedelta(days=1),
                           periods=horizon, freq='D')

plt.figure(figsize=(10,4))
plt.plot(res.conditional_volatility.index,
         res.conditional_volatility,
         label='σ_t (historic)')
plt.plot(future_idx,
         std_fore.values,
         '--', label='σ_{T+h} forecast')
plt.fill_between(future_idx,
                 std_fore.values * 0.95,
                 std_fore.values * 1.05,
                 alpha=0.2,
                 label='±5% band')
plt.title('GARCH(1,1) Volatility Forecast (5 days)')
plt.legend()
plt.show()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from typing import List
import statsmodels.api as sm


def import_sql_dataset(database_path: str, table_name: str) -> pd.DataFrame:
    """
    Import dataset from a local SQLite database.

    Args:
        database_path (str): Path to the SQLite database file.
        table_name (str): Name of the table to import.

    Returns:
        pd.DataFrame: Dataset loaded from the SQLite table.
    """
    if not Path(database_path).exists():
        raise FileNotFoundError(f"Database file not found: {database_path}")

    conn = sqlite3.connect(database_path)
    try:
        df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    finally:
        conn.close()
    return df


def load_and_filter_fx_swap_data(
    database_path: str,
    table_name: str,
    ticker_list: List[str],
    start_date: str,
    end_date: str,
    product_type: str,
    near_bucket: str,
    far_bucket: str
) -> pd.DataFrame:
    """
    Loads and filters FX Swap data from a SQL database.

    Returns:
        pd.DataFrame: Filtered and sorted DataFrame.
    """
    df = import_sql_dataset(database_path, table_name)

    filtered_df = df[
        df['sym'].isin(ticker_list) &
        (df['near_leg_bucket'] == near_bucket) &
        (df['far_leg_bucket'] == far_bucket) &
        (df['product_type'] == product_type) &
        (df['date'] >= start_date) &
        (df['date'] <= end_date)
    ]

    filtered_df['date'] = pd.to_datetime(filtered_df['date'])
    filtered_df = filtered_df.sort_values('date')

    return filtered_df


def decompose_and_plot(series: pd.Series):
    """
    Decomposes the time series to extract trend and seasonality components.
    """
    decomposition = sm.tsa.seasonal_decompose(series, model='additive', period=30)
    plt.figure(figsize=(12, 8))
    decomposition.plot()
    plt.tight_layout()
    plt.show()
    return decomposition


def main():
    database_path = '/Users/santnol/soda/storage/volume_model/fxf/rfq/daily/fxf_rfq_volume_daily.db'
    table_name = 'G11'

    # Filter and load FX Swap data
    filtered_df = load_and_filter_fx_swap_data(
        database_path=database_path,
        table_name=table_name,
        ticker_list=['EURUSD'],
        start_date='2025-01-01',
        end_date='2025-06-30',
        product_type='FX Outright',
        near_bucket='PRE_SPOT',
        far_bucket='PRE_SPOT'
    )

    # Print basic statistics
    print(f"Dataset shape: {filtered_df.shape}")
    print(f"Date range: {filtered_df['date'].min()} to {filtered_df['date'].max()}")
    print(f"Total volume: ${filtered_df['req_vol_usd'].sum():,.2f}")

    # Plot the request volume over time
    plt.figure(figsize=(12, 6))
    plt.plot(filtered_df['date'], filtered_df['req_vol_usd'], marker='o', linewidth=2, markersize=4)
    plt.title('EURUSD FX Swap Volume Over Time (PRE_SPOT - PRE_SPOT)')
    plt.xlabel('Date')
    plt.ylabel('Request Volume (USD)')
    plt.grid(visible=True, alpha=0.3)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Optional: Decompose time series
    # decompose_and_plot(filtered_df.set_index('date')['req_vol_usd'])


if __name__ == "__main__":
    main()
```
