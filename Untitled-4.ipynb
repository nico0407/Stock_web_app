\section{Introduction}
We seek to decompose a daily time series $X_t$ (``volume'') into three parts: the trend component $T_t$, the seasonal component $S_t$, and the residual~$R_t$.  The additive model is
\begin{align}
X_t = T_t + S_t + R_t.
\end{align}

\section{Trend Estimation via Moving Average}
A simple estimate of the trend is the centered moving average (MA) of window length $m=2k+1$.  Define
\begin{align}
\widehat{T}_t = \frac{1}{m}\sum_{i=-k}^{k} X_{t+i},
\end{align}
where $m=30$ days in our example ($k=14$).  In Python:
# 30-day centered moving average
df['trend'] = df['volume'].rolling(window=30, center=True).mean()
# detrended series:
df['detrended'] = df['volume'] - df['trend']

\section{Classical Seasonal Decomposition}
Given a period $m$ (here weekly, $m=7$), classical additive decomposition estimates $T_t$ by smoothing, then the seasonal component by averaging:
\begin{align}
\widehat{T}_t &= \text{MA}(X_t),\\
\widehat{S}_j &= \frac{1}{N}\sum_{\substack{t=1\dots,N\\t\equiv j\,(\mathrm{mod}\,m)}} \bigl(X_t - \widehat{T}_t\bigr),\quad j=1,\dots,m,\\
\widehat{R}_t &= X_t - \widehat{T}_t - \widehat{S}_{t\bmod m}.
\end{align}
In Python with \texttt{statsmodels}:
from statsmodels.tsa.seasonal import seasonal_decompose
decomp = seasonal_decompose(df['volume'], model='additive', period=7)
decomp.plot()


\section{STL Decomposition}
STL (\emph{Seasonal and Trend via Loess}) fits both components via local regression.  It handles irregular seasonal patterns and outliers:
from statsmodels.tsa.seasonal import STL
stl = STL(df['volume'], period=7, robust=True)
res = stl.fit()
res.plot()

\section{Autocorrelation Analysis}
Seasonality is confirmed by spikes in the autocorrelation function (ACF) at lags that are multiples of~$m$:
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df['volume'].dropna(), lags=30)

\section{Summary of Key Formulas}
\begin{align}
X_t &= T_t + S_t + R_t,\\
\widehat{T}_t &= \frac{1}{m}\sum_{i=-k}^{k} X_{t+i},\\
\widehat{S}_j &= \frac{1}{N}\sum_{t\equiv j\,(\mathrm{mod}\,m)} \bigl(X_t - \widehat{T}_t\bigr),\\
\widehat{R}_t &= X_t - \widehat{T}_t - \widehat{S}_t.
\end{align}

\section{GARCH(1,1) Modeling of Daily Volume}

We start from the raw volume series \(X_t\) and form the logâ€‘returns
\begin{align}
r_t = \ln(X_t) - \ln(X_{t-1}),\quad t=2,\dots,T.
\end{align}

\subsection{Stationarity Check}
Perform the Augmented Dickeyâ€“Fuller test on \(\{r_t\}\):
\begin{align}
H_0:\ \text{unit root (nonstationary)}
\quad\text{vs.}\quad
H_1:\ \text{stationary}.
\end{align}
Reject \(H_0\) if the ADF \(p\)-value \(<0.05\).

\subsection{GARCH(1,1) Specification}
Model the returns via
\begin{align}
r_t &= \mu + \varepsilon_t,\quad
\varepsilon_t = \sigma_t\,z_t,\quad z_t\sim \mathcal{N}(0,1),\\
\sigma_t^2 &= \omega + \alpha\,\varepsilon_{t-1}^2 + \beta\,\sigma_{t-1}^2,
\end{align}
with parameter constraints for covariance stationarity:
\begin{align}
\omega>0,\quad \alpha\ge0,\quad \beta\ge0,\quad \alpha + \beta < 1.
\end{align}

\subsection{Unconditional Variance}
Under stationarity,
\begin{align}
\operatorname{Var}(r_t)
= \mathbb{E}\bigl[\sigma_t^2\bigr]
= \frac{\omega}{1 - \alpha - \beta}.
\end{align}

\subsection{Diagnostics}
After fitting, extract standardized residuals
\(\displaystyle z_t = \varepsilon_t/\sigma_t\)
and check:1) ACF of \(z_t\) and \(z_t^2\) (no significant lags). 2) tem Ljungâ€“Box tests on \(z_t\) and \(z_t^2\) (\(p\)-values \(>0.05\)).


# ðŸ“˜ Deep Explanation of the GARCH Model for Time Series

---

## ðŸ”¹ Introduction

The **GARCH model** (Generalized Autoregressive Conditional Heteroskedasticity) is a widely used tool in time series analysisâ€”especially in financial econometricsâ€”for modeling and forecasting **conditional variance** (volatility). It generalizes the ARCH model introduced by *Engle (1982)* and was extended to GARCH by *Bollerslev (1986)*.

---

## ðŸ”¹ Objective

The goal is to model a **time-varying variance** in time series data, especially when there is evidence of **volatility clustering** (i.e., large changes tend to follow large changes, and small changes follow small changes).

---

## ðŸ”¹ Model Framework

Let \( r_t \) denote the return (or a general time series variable). We assume:

$$
r_t = \mu + \epsilon_t
$$

- \( \mu \): conditional mean (can be constant or ARMA process)  
- \( \epsilon_t \): innovation or shock term

We model \( \epsilon_t \) as:

$$
\epsilon_t = \sigma_t z_t
$$

Where:
- \( z_t \sim \text{i.i.d. } (0, 1) \), usually standard normal  
- \( \sigma_t^2 \): conditional variance

---

## ðŸ”¹ ARCH(q) Model (Engle, 1982)

The ARCH(q) model defines conditional variance as:

$$
\sigma_t^2 = \alpha_0 + \sum_{i=1}^{q} \alpha_i \epsilon_{t-i}^2
$$

- \( \alpha_0 > 0 \), \( \alpha_i \geq 0 \)

---

## ðŸ”¹ GARCH(p, q) Model (Bollerslev, 1986)

The GARCH model extends ARCH by including lagged variances:

$$
\sigma_t^2 = \alpha_0 + \sum_{i=1}^{q} \alpha_i \epsilon_{t-i}^2 + \sum_{j=1}^{p} \beta_j \sigma_{t-j}^2
$$

- \( \alpha_0 > 0 \), \( \alpha_i, \beta_j \geq 0 \)  
- Stationarity requires: \( \sum \alpha_i + \sum \beta_j < 1 \)

---

## ðŸ”¹ GARCH(1,1): Most Common Case

$$
\sigma_t^2 = \alpha_0 + \alpha_1 \epsilon_{t-1}^2 + \beta_1 \sigma_{t-1}^2
$$

**Interpretation**:
- \( \alpha_0 \): Long-run average variance  
- \( \alpha_1 \): Response to recent news (ARCH effect)  
- \( \beta_1 \): Persistence of volatility (GARCH effect)

---

## ðŸ”¹ Conditions for GARCH(1,1)

- **Positivity**: \( \alpha_0 > 0, \alpha_1 \geq 0, \beta_1 \geq 0 \)  
- **Stationarity**: \( \alpha_1 + \beta_1 < 1 \)  
- **Unconditional variance**:

$$
\mathbb{E}[\sigma_t^2] = \frac{\alpha_0}{1 - \alpha_1 - \beta_1}
$$

---

## ðŸ”¹ Log-Likelihood Function (Gaussian Case)

Assuming \( z_t \sim \mathcal{N}(0,1) \), the log-likelihood for \( T \) observations is:

$$
\mathcal{L}(\theta) = -\frac{T}{2} \log(2\pi) - \frac{1}{2} \sum_{t=1}^{T} \left( \log(\sigma_t^2) + \frac{\epsilon_t^2}{\sigma_t^2} \right)
$$

Where \( \theta = (\mu, \alpha_0, \alpha_1, \beta_1) \)

---

## ðŸ”¹ Forecasting with GARCH(1,1)

**1-step ahead forecast**:

$$
\hat{\sigma}_{t+1}^2 = \alpha_0 + \alpha_1 \epsilon_t^2 + \beta_1 \sigma_t^2
$$

**h-step ahead forecast**:

As \( h \to \infty \):

$$
\hat{\sigma}_{t+h}^2 \to \frac{\alpha_0}{1 - \alpha_1 - \beta_1}
$$

---

## ðŸ”¹ Model Properties

- **Volatility Clustering**: High-volatility periods cluster together  
- **Leptokurtosis**: Even if \( z_t \) is normal, \( r_t \) has fat tails  
- **Persistence**: High \( \alpha_1 + \beta_1 \) implies long memory

---

## ðŸ”¹ GARCH Extensions

- **EGARCH**: Handles asymmetry (leverage effect)  
- **GJR-GARCH**: Adds dummy for negative shocks  
- **MGARCH**: Multivariate GARCH  
- **IGARCH**: Integrated GARCH with \( \alpha_1 + \beta_1 = 1 \)

---

## ðŸ”¹ Summary Table

| Term              | Meaning                            |
|-------------------|-------------------------------------|
| \( \alpha_0 \)    | Constant term (long-run variance)   |
| \( \alpha_i \)    | Reaction to recent shocks (ARCH)    |
| \( \beta_j \)     | Persistence of volatility (GARCH)   |
| \( \epsilon_t \)  | Innovation/shock                    |
| \( \sigma_t^2 \)  | Conditional variance                |
| \( z_t \)         | Standardized shock                  |

---

# ðŸ“˜ In-Depth GARCH Model for Time Series Analysis

---

## ðŸ”¹ Introduction

The **GARCH model** (Generalized Autoregressive Conditional Heteroskedasticity) is a statistical model used to estimate **time-varying volatility** in time series, particularly financial returns. It was developed as an extension of the **ARCH** model by Engle (1982) and generalized by Bollerslev (1986).

---

## ðŸ”¹ Model Framework

Let \( r_t \) be a return series (or residual from a mean model):

\[
r_t = \mu + \epsilon_t \quad \text{with} \quad \epsilon_t = \sigma_t z_t
\]

Where:
- \( \mu \): conditional mean (could be a constant or ARMA model)
- \( z_t \sim \text{i.i.d. } (0, 1) \)
- \( \sigma_t^2 \): conditional variance (the focus of GARCH)

---

## ðŸ”¹ ARCH(q) Model (Engle, 1982)

\[
\sigma_t^2 = \alpha_0 + \sum_{i=1}^{q} \alpha_i \epsilon_{t-i}^2
\]

- \( \alpha_0 > 0 \), \( \alpha_i \geq 0 \)

---

## ðŸ”¹ GARCH(p, q) Model (Bollerslev, 1986)

\[
\sigma_t^2 = \alpha_0 + \sum_{i=1}^{q} \alpha_i \epsilon_{t-i}^2 + \sum_{j=1}^{p} \beta_j \sigma_{t-j}^2
\]

- \( \alpha_0 > 0 \), \( \alpha_i, \beta_j \geq 0 \)
- Stationarity condition: \( \sum \alpha_i + \sum \beta_j < 1 \)

---

## ðŸ”¹ GARCH(1,1) â€“ The Workhorse

\[
\sigma_t^2 = \alpha_0 + \alpha_1 \epsilon_{t-1}^2 + \beta_1 \sigma_{t-1}^2
\]

**Interpretation**:
- \( \alpha_0 \): baseline variance
- \( \alpha_1 \): impact of past shock (ARCH effect)
- \( \beta_1 \): volatility persistence (GARCH effect)

---

## ðŸ”¹ Stationarity and Moments

### Unconditional Variance:

\[
\mathbb{E}[\sigma_t^2] = \frac{\alpha_0}{1 - \alpha_1 - \beta_1}
\]

Valid if: \( \alpha_1 + \beta_1 < 1 \)

### Kurtosis (Fat Tails):

Even with normal \( z_t \), GARCH processes can show **leptokurtosis** (fat tails). The 4th moment may diverge if parameters are too close to 1.

---

## ðŸ”¹ Likelihood Function (Gaussian Innovations)

For \( T \) observations and assuming \( z_t \sim \mathcal{N}(0,1) \):

\[
\mathcal{L}(\theta) = -\frac{T}{2} \log(2\pi) - \frac{1}{2} \sum_{t=1}^{T} \left[ \log(\sigma_t^2) + \frac{\epsilon_t^2}{\sigma_t^2} \right]
\]

---

## ðŸ”¹ Forecasting with GARCH(1,1)

### One-step ahead:

\[
\hat{\sigma}_{t+1}^2 = \alpha_0 + \alpha_1 \epsilon_t^2 + \beta_1 \sigma_t^2
\]

### h-step ahead:

\[
\hat{\sigma}_{t+h}^2 \to \frac{\alpha_0}{1 - \alpha_1 - \beta_1} \quad \text{as } h \to \infty
\]

---

## ðŸ”¹ When to Use Which GARCH Model

| Model         | When to Use                                                                 |
|---------------|------------------------------------------------------------------------------|
| ARCH(q)       | Only short memory in volatility; simple and rarely optimal alone            |
| GARCH(1,1)    | Symmetric volatility clustering, works well in practice                     |
| EGARCH        | Volatility responds asymmetrically to shocks (e.g., leverage effects)       |
| GJR-GARCH     | Negative shocks impact volatility more than positive ones (threshold effect)|
| IGARCH        | High persistence; variance is non-stationary (unit root in volatility)      |
| TGARCH        | Threshold effects, nonlinear response to shocks                             |
| APARCH        | Asymmetric Power ARCHâ€”flexible modeling, includes many GARCH variants       |

---

## ðŸ”¹ How to Choose the Best Model

1. **Visual Checks**
   - Plot returns and squared returns.
   - Look for volatility clustering.

2. **ARCH Effects Test**
   - Use ARCH LM test to confirm heteroskedasticity.

3. **Fit Base Models**
   - Start with GARCH(1,1).
   - Check log-likelihood, AIC, BIC.

4. **Test for Asymmetry**
   - Try EGARCH or GJR-GARCH.

5. **Compare Models**
   - AIC / BIC / log-likelihood.
   - Residual diagnostics (Ljung-Box on squared residuals).
   - Out-of-sample forecast performance.

# ðŸ“˜ Advanced GARCH Topics: Multivariate GARCH & Volatility Forecast Evaluation

---

## ðŸ”¹ Multivariate GARCH (MGARCH)

Multivariate GARCH models allow for time-varying **covariances** and **correlations** between multiple return series. This is crucial in:

- Portfolio risk modeling
- Systemic risk assessment
- Dynamic hedging and asset pricing

### ðŸ”¸ General Setup

Let:

\[
\mathbf{r}_t = \boldsymbol{\mu}_t + \boldsymbol{\epsilon}_t, \quad \boldsymbol{\epsilon}_t \sim \mathcal{N}(\mathbf{0}, \mathbf{H}_t)
\]

- \( \mathbf{r}_t \in \mathbb{R}^n \): returns
- \( \mathbf{H}_t \): \( n \times n \) time-varying covariance matrix
- Goal: model the evolution of \( \mathbf{H}_t \)

---

### ðŸ”¸ Key MGARCH Models

| Model        | Description                                                                 | Scalability |
|--------------|-----------------------------------------------------------------------------|-------------|
| **VECH**     | Models vectorized lower triangle of \( \mathbf{H}_t \)                      | Poor        |
| **BEKK**     | Guarantees positive definiteness                                            | Moderate    |
| **DCC**      | Decomposes volatilities and correlations; most popular                      | Good        |
| **CCC**      | Constant correlation, varying variances                                     | Good        |

---

### ðŸ”¸ DCC-GARCH (Dynamic Conditional Correlation)

DCC-GARCH separates variance and correlation dynamics:

\[
\mathbf{H}_t = \mathbf{D}_t \mathbf{R}_t \mathbf{D}_t
\]

Where:
- \( \mathbf{D}_t = \text{diag}(\sigma_{1t}, \ldots, \sigma_{nt}) \)
- \( \mathbf{R}_t \): time-varying correlation matrix

Update equation for correlation matrix:

\[
Q_t = (1 - a - b)\bar{Q} + a z_{t-1} z_{t-1}^\top + b Q_{t-1}
\]
\[
R_t = \text{diag}(Q_t)^{-1/2} Q_t \text{diag}(Q_t)^{-1/2}
\]

---

## ðŸ”¹ Volatility Forecast Evaluation

After fitting a GARCH model, you must assess how accurate its **volatility forecasts** are.

---

### ðŸ”¸ Step 1: Forecast Volatility

Using a GARCH(1,1) model:

\[
\hat{\sigma}_{t+1}^2 = \alpha_0 + \alpha_1 \epsilon_t^2 + \beta_1 \sigma_t^2
\]

---

### ðŸ”¸ Step 2: Choose an Evaluation Metric

#### âœ… Scale-Dependent Metrics

- **MSE** (Mean Squared Error):
  \[
  \text{MSE} = \frac{1}{T} \sum_{t=1}^T (r_t^2 - \hat{\sigma}_t^2)^2
  \]

- **MAE** (Mean Absolute Error):
  \[
  \text{MAE} = \frac{1}{T} \sum_{t=1}^T |r_t^2 - \hat{\sigma}_t^2|
  \]

#### âœ… Proper Loss Functions

- **QLIKE** (Quasi-Likelihood):
  \[
  \text{QLIKE} = \frac{1}{T} \sum_{t=1}^T \left[ \log(\hat{\sigma}_t^2) + \frac{r_t^2}{\hat{\sigma}_t^2} \right]
  \]

- **Logarithmic Loss**:
  \[
  \text{LogLoss} = -\log \left( \frac{1}{\sqrt{2\pi \hat{\sigma}_t^2}} \exp\left(-\frac{r_t^2}{2\hat{\sigma}_t^2} \right) \right)
  \]

---

### ðŸ”¸ Step 3: Compare Models â€” Diebold-Mariano Test

The **Diebold-Mariano (DM)** test evaluates whether two forecasting models have significantly different accuracy.

\[
DM = \frac{\bar{d}}{\sqrt{\frac{2\pi \hat{f}_d(0)}{T}}}
\]

- \( \bar{d} \): mean loss differential between two models
- \( \hat{f}_d(0) \): long-run variance of loss differential

---

### ðŸ”¸ Step 4: Backtesting Value-at-Risk (VaR)

If using GARCH for VaR forecasting:

1. Compute:
   \[
   \text{VaR}_{\alpha,t} = \hat{\mu}_t + \hat{\sigma}_t z_\alpha
   \]

2. Count exceedances:
   \[
   I_t = 
   \begin{cases}
   1 & r_t < \text{VaR}_{\alpha,t} \\
   0 & \text{otherwise}
   \end{cases}
   \]

3. Run tests:
   - **Kupiec test**: unconditional coverage
   - **Christoffersen test**: independence of exceedances

## ðŸ”¹ 1. Understand the Data First

Before choosing a GARCH model, analyze the **time series characteristics**:

| Feature                      | What to Check                                                   |
|-----------------------------|------------------------------------------------------------------|
| Volatility Clustering       | Are large changes grouped over time?                            |
| Leverage Effects            | Do negative returns increase volatility more than positives?    |
| Fat Tails                   | Do large returns occur more frequently than in a normal dist.?  |
| Structural Breaks           | Are there regime shifts (e.g., crisis periods)?                 |
| Long Memory                 | Does volatility decay slowly (persistent effects)?              |
| Multiple Assets             | Are correlations between assets important?                      |

## ðŸ”¹ 2. General Workflow

1. **Model the Mean**  
   Fit ARIMA/ARMA or use a constant mean.

2. **Check for ARCH Effects**  
   Use **Engleâ€™s ARCH LM test** on residuals. If significant â†’ proceed to GARCH modeling.

3. **Start Simple**  
   Use **GARCH(1,1)** as a baseline.

---

## ðŸ”¹ 3. Recommended GARCH Variants

| Scenario                               | Use This Model               | Why?                                         |
|----------------------------------------|-------------------------------|----------------------------------------------|
| Basic volatility clustering            | GARCH(1,1)                    | Simple, effective, widely used               |
| Asymmetric volatility (leverage)       | EGARCH or GJR-GARCH           | Captures impact of negative shocks           |
| High persistence (\( \alpha+\beta \approx 1 \)) | IGARCH                        | Unit root in variance                        |
| Long-memory volatility                 | FIGARCH                       | Slow decay of shocks                         |
| Portfolio of assets                    | DCC-GARCH (or MGARCH)         | Time-varying covariances                     |
| Frequent regime shifts                 | Markov-switching GARCH        | Captures volatility regimes                  |
| Fat tails or skewness                  | GARCH + t or skewed-t         | Heavy-tailed distributions                   |

## ðŸ”¹ 4. Model Selection Criteria

After fitting models, compare using:

| Criterion               | Goal                         |
|-------------------------|------------------------------|
| **AIC / BIC**           | Lower is better              |
| **Log-likelihood**      | Higher is better             |
| **QLIKE / MSE / MAE**   | Better forecast fit          |
| **Residual Tests**      | No remaining ARCH effects    |
| **DM Test**             | Compares forecast accuracy   |

## ðŸ”¹ 6. Forecast Comparison and Evaluation

1. Generate volatility forecasts:
   \[
   \hat{\sigma}_{t+1}^2 = \alpha_0 + \alpha_1 \epsilon_t^2 + \beta_1 \sigma_t^2
   \]

2. Evaluate using:

- **QLIKE**:
  \[
  \text{QLIKE} = \sum \left[ \log(\hat{\sigma}_t^2) + \frac{r_t^2}{\hat{\sigma}_t^2} \right]
  \]

- **Diebold-Mariano test** to compare forecast accuracy of two models

- Use rolling estimation for evolving markets

- Compare models with both in-sample fit and out-of-sample forecast

````python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from typing import List
import statsmodels.api as sm


def import_sql_dataset(database_path: str, table_name: str) -> pd.DataFrame:
    """
    Import dataset from a local SQLite database.

    Args:
        database_path (str): Path to the SQLite database file.
        table_name (str): Name of the table to import.

    Returns:
        pd.DataFrame: Dataset loaded from the SQLite table.
    """
    if not Path(database_path).exists():
        raise FileNotFoundError(f"Database file not found: {database_path}")

    conn = sqlite3.connect(database_path)
    try:
        df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    finally:
        conn.close()
    return df


def load_and_filter_fx_swap_data(
    database_path: str,
    table_name: str,
    ticker_list: List[str],
    start_date: str,
    end_date: str,
    product_type: str,
    near_bucket: str,
    far_bucket: str
) -> pd.DataFrame:
    """
    Loads and filters FX Swap data from a SQL database.

    Returns:
        pd.DataFrame: Filtered and sorted DataFrame.
    """
    df = import_sql_dataset(database_path, table_name)

    filtered_df = df[
        df['sym'].isin(ticker_list) &
        (df['near_leg_bucket'] == near_bucket) &
        (df['far_leg_bucket'] == far_bucket) &
        (df['product_type'] == product_type) &
        (df['date'] >= start_date) &
        (df['date'] <= end_date)
    ]

    filtered_df['date'] = pd.to_datetime(filtered_df['date'])
    filtered_df = filtered_df.sort_values('date')

    return filtered_df


def decompose_and_plot(series: pd.Series):
    """
    Decomposes the time series to extract trend and seasonality components.
    """
    decomposition = sm.tsa.seasonal_decompose(series, model='additive', period=30)
    plt.figure(figsize=(12, 8))
    decomposition.plot()
    plt.tight_layout()
    plt.show()
    return decomposition


def main():
    database_path = '/Users/santnol/soda/storage/volume_model/fxf/rfq/daily/fxf_rfq_volume_daily.db'
    table_name = 'G11'

    # Filter and load FX Swap data
    filtered_df = load_and_filter_fx_swap_data(
        database_path=database_path,
        table_name=table_name,
        ticker_list=['EURUSD'],
        start_date='2025-01-01',
        end_date='2025-06-30',
        product_type='FX Outright',
        near_bucket='PRE_SPOT',
        far_bucket='PRE_SPOT'
    )

    # Print basic statistics
    print(f"Dataset shape: {filtered_df.shape}")
    print(f"Date range: {filtered_df['date'].min()} to {filtered_df['date'].max()}")
    print(f"Total volume: ${filtered_df['req_vol_usd'].sum():,.2f}")

    # Plot the request volume over time
    plt.figure(figsize=(12, 6))
    plt.plot(filtered_df['date'], filtered_df['req_vol_usd'], marker='o', linewidth=2, markersize=4)
    plt.title('EURUSD FX Swap Volume Over Time (PRE_SPOT - PRE_SPOT)')
    plt.xlabel('Date')
    plt.ylabel('Request Volume (USD)')
    plt.grid(visible=True, alpha=0.3)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Optional: Decompose time series
    # decompose_and_plot(filtered_df.set_index('date')['req_vol_usd'])


if __name__ == "__main__":
    main()
````
