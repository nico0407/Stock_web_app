```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, kpss, acf, pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import het_arch
from arch import arch_model
import seaborn as sns


# Generate 10 sequential dates starting from today
dates = [datetime.today() - timedelta(days=i) for i in range(365)]

# Generate random volume values
volume1 = np.random.randint(100, 1000, size=365)
volume2 = np.random.randint(100, 1000, size=365)
volume3 = np.random.randint(100, 1000, size=365)
volume4 = np.random.randint(100, 1000, size=365)

# Create the DataFrame
FX_Data = pd.DataFrame({
    'date': dates,
    'volume1': volume1,
    'volume2': volume2,
    'volume3': volume3,
    'volume4': volume4,
})

FX_Data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.api import VAR
from arch import arch_model
import seaborn as sns

# Load and preprocess
FX_Data['date'] = pd.to_datetime(FX_Data['date'])
FX_Data.set_index('date', inplace=True)
FX_Data.sort_index(inplace=True)

# Subset of volume columns
volumes = FX_Data[['volume1', 'volume2', 'volume3', 'volume4']]

# ========== 1. Covariation Analysis ==========
print("Covariance matrix:")
print(volumes.cov())

print("\nCorrelation matrix:")
print(volumes.corr())

sns.heatmap(volumes.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Volumes")
plt.show()

# ========== 2. Cointegration Test (Johansen) ==========
# Johansen Cointegration Test
def johansen_cointegration_test(df, det_order=-1, k_ar_diff=1):
    result = coint_johansen(df, det_order, k_ar_diff)
    trace_stat = result.lr1
    critical_values = result.cvt
    for i in range(len(trace_stat)):
        print(f"Rank {i}: Trace Statistic = {trace_stat[i]:.4f}, Critical Values (90%, 95%, 99%) = {critical_values[i]}")

print("\nJohansen Cointegration Test:")
johansen_cointegration_test(volumes)

# ========== 3. VAR Model and State Space ==========

# First difference to make stationary if needed
diff_volumes = volumes.diff().dropna()

# Fit Vector Autoregression
model = VAR(diff_volumes)
results = model.fit(maxlags=5, ic='aic')
print(results.summary())

# State space representation with VARMAX (for volatility, we augment later)
state_model = VARMAX(diff_volumes, order=(1,0))
state_fit = state_model.fit(disp=False)
print(state_fit.summary())

# ========== 4. Volatility Modeling (ARCH/GARCH per series) ==========

for col in volumes.columns:
    print(f"\nFitting GARCH(1,1) model to residuals of {col}")
    series = diff_volumes[col].dropna()
    garch = arch_model(series, vol='Garch', p=1, q=1)
    garch_fitted = garch.fit(disp='off')
    print(garch_fitted.summary())
    garch_fitted.plot()
    plt.title(f'GARCH Volatility for {col}')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, coint
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.statespace.varmax import VARMAX
#from statsmodels.tsa.statespace.tools import cfa
from arch import arch_model
import seaborn as sns
from statsmodels.tsa.api import VECM
from statsmodels.stats.moment_helpers import corr2cov

# Set style for plots
#plt.style.use('seaborn')
sns.set_palette("husl")

# Assuming FX_Data is already loaded as a pandas DataFrame
# If not, you would load it with:
# FX_Data = pd.read_csv('your_file.csv', parse_dates=['date'], index_col='date')

# Ensure date is the index and sorted
FX_Data = FX_Data.set_index('date').sort_index()

# Basic data inspection
print(FX_Data.head())
print("\nData info:")
print(FX_Data.info())
print("\nSummary statistics:")
print(FX_Data.describe())

## 1. Covariation Analysis
# Correlation matrix
corr_matrix = FX_Data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Volume Series')
plt.show()

# Rolling correlation (example with volume1 and volume2)
window_size = 20  # 20 trading days
FX_Data['rolling_corr'] = FX_Data['volume1'].rolling(window=window_size).corr(FX_Data['volume2'])
plt.figure(figsize=(12, 6))
FX_Data['rolling_corr'].plot(title=f'Rolling Correlation (Window={window_size} days) between Volume1 and Volume2')
plt.ylabel('Correlation Coefficient')
plt.grid(True)
plt.show()

# Cross-correlation analysis
from statsmodels.tsa.stattools import ccf
max_lag = 20
ccf_values = ccf(FX_Data['volume1'].dropna(), FX_Data['volume2'].dropna(), unbiased=False)
plt.figure(figsize=(12, 6))
plt.stem(range(-max_lag, max_lag+1), ccf_values[:2*max_lag+1])
plt.title('Cross-Correlation Function (Volume1 vs Volume2)')
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.grid(True)
plt.show()

## 2. Cointegration Analysis
# Perform pairwise cointegration tests
volume_columns = ['volume1', 'volume2', 'volume3', 'volume4']
coint_results = []

for i in range(len(volume_columns)):
    for j in range(i+1, len(volume_columns)):
        series1 = FX_Data[volume_columns[i]]
        series2 = FX_Data[volume_columns[j]]
        score, pvalue, _ = coint(series1, series2)
        coint_results.append({
            'Series1': volume_columns[i],
            'Series2': volume_columns[j],
            'Test Statistic': score,
            'p-value': pvalue
        })

coint_df = pd.DataFrame(coint_results)
print("\nCointegration Test Results:")
print(coint_df)

# Plot potential cointegrated pairs (example with first pair)
if len(coint_df) > 0:
    best_pair = coint_df.loc[coint_df['p-value'].idxmin()]
    s1 = FX_Data[best_pair['Series1']]
    s2 = FX_Data[best_pair['Series2']]
    
    plt.figure(figsize=(12, 6))
    plt.plot(s1, label=best_pair['Series1'])
    plt.plot(s2, label=best_pair['Series2'])
    plt.title(f"Most Significant Cointegrated Pair: {best_pair['Series1']} vs {best_pair['Series2']}")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Plot spread of cointegrated pair
    spread = s1 - s2
    plt.figure(figsize=(12, 6))
    spread.plot(title=f"Spread between {best_pair['Series1']} and {best_pair['Series2']}")
    plt.axhline(spread.mean(), color='red', linestyle='--')
    plt.grid(True)
    plt.show()

## 3. Vector Autoregression (VAR) Modeling
# Check stationarity of all series first
for col in volume_columns:
    adf_result = adfuller(FX_Data[col].dropna())
    print(f"\nADF Test for {col}:")
    print(f"  ADF Statistic: {adf_result[0]}")
    print(f"  p-value: {adf_result[1]}")

# Difference if necessary (example)
FX_Data_diff = FX_Data[volume_columns].diff().dropna()

# Determine optimal VAR lag order
var_model = VAR(FX_Data_diff)
lag_results = var_model.select_order(maxlags=10)
print("\nVAR Lag Selection:")
print(lag_results.summary())

# Fit VAR model with selected lags
selected_lags = lag_results.aic  # Using AIC criterion
var_results = var_model.fit(maxlags=selected_lags)
print("\nVAR Model Results:")
print(var_results.summary())

# Granger Causality tests with error handling
print("\nGranger Causality Tests:")
for col in volume_columns:
    try:
        print(f"\nTesting if other variables Granger-cause {col}:")
        other_cols = [c for c in volume_columns if c != col]
        if len(other_cols) > 0:  # Ensure we have columns to test against
            granger_test = var_results.test_causality(col, other_cols, kind='f')
            print(granger_test.summary())
        else:
            print("No other variables to test against")
    except Exception as e:
        print(f"Error testing causality for {col}: {str(e)}")

# Impulse Response Analysis
irf = var_results.irf(10)
irf.plot(orth=True, figsize=(12, 10))
plt.suptitle('Impulse Response Functions')
plt.tight_layout()
plt.show()

## 4. State Space Modeling with Volatility Processes
# VARMAX model (VAR + MA terms in state space framework)
# Using differenced data if needed
varmax_model = VARMAX(FX_Data_diff, order=(1, 1))  # VAR(1) + MA(1)
varmax_results = varmax_model.fit(disp=False)
print("\nVARMAX Model Results:")
print(varmax_results.summary())

# Extract residuals for volatility modeling
residuals = varmax_results.resid

# Multivariate GARCH modeling (DCC-GARCH)
from arch import arch_model
from arch.univariate import GARCH, EWMAVariance

# Fit univariate GARCH models for each series
garch_models = {}
for col in residuals.columns:
    garch = arch_model(residuals[col], mean='Zero', vol='Garch', p=1, q=1)
    garch_fit = garch.fit(update_freq=5, disp='off')
    garch_models[col] = garch_fit
    print(f"\nGARCH Model for {col}:")
    print(garch_fit.summary())

# Plot conditional volatilities
plt.figure(figsize=(12, 8))
for col in residuals.columns:
    plt.plot(garch_models[col].conditional_volatility, label=col)
plt.title('Conditional Volatility from GARCH Models')
plt.legend()
plt.grid(True)
plt.show()

# Dynamic Conditional Correlation (DCC) model
from arch import arch_model
from arch.univariate import GARCH, EWMAVariance
from arch.covariance import DCC

# Need to reshape residuals for DCC
returns = residuals.values

# Create DCC model
dcc = DCC(returns)
dcc_results = dcc.fit(update_freq=5)
print("\nDCC Model Results:")
print(dcc_results.summary())

# Plot dynamic correlations
correlations = dcc_results.dynamic_correlation
fig, axes = plt.subplots(len(volume_columns), len(volume_columns), figsize=(15, 15))
for i in range(len(volume_columns)):
    for j in range(len(volume_columns)):
        if i == j:
            axes[i,j].plot(dcc_results.conditional_volatility[:,i])
            axes[i,j].set_title(f'Volatility {volume_columns[i]}')
        elif i < j:
            axes[i,j].plot(correlations[:,i,j])
            axes[i,j].set_title(f'Corr {volume_columns[i]}-{volume_columns[j]}')
        else:
            axes[i,j].axis('off')
plt.tight_layout()
plt.show()

## 5. VECM for Cointegrated Series (if cointegration found)
if len(coint_df[coint_df['p-value'] < 0.05]) > 0:
    print("\nEstimating VECM for cointegrated series:")
    # Using the most significant cointegrated pair
    vecm_data = FX_Data[[best_pair['Series1'], best_pair['Series2']]].dropna()
    vecm_model = VECM(vecm_data, k_ar_diff=2, coint_rank=1)
    vecm_results = vecm_model.fit()
    print(vecm_results.summary())
    
    # Plot the cointegration relation
    plt.figure(figsize=(12, 6))
    plt.plot(vecm_results.coint)
    plt.title('Cointegration Relation')
    plt.axhline(0, color='red', linestyle='--')
    plt.grid(True)
    plt.show()
```
